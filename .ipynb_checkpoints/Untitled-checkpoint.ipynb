```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import PIL
import pathlib
from pathlib import Path
import os
import seaborn as sns
print (tf.__version__)
```

```python
directory = 'MathSymbols'
files = Path(directory).glob('*')
for dir_path in files:
    print(dir_path,len([entry for entry in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, entry))]))
    
```

```python
tf.keras.utils.image_dataset_from_directory(
    'MathSymbols',
    labels='inferred',
    label_mode='int',
    class_names=None,
    color_mode='grayscale',
    batch_size=32,
    image_size=(45, 45),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
)
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
  'MathSymbols',
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(45, 45),
  batch_size=32)
val_ds = tf.keras.utils.image_dataset_from_directory(
  'MathSymbols',
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(45, 45),
  batch_size=32)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(16):
    ax = plt.subplot(4, 4, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)
normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
num_classes = 82

model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(num_classes)
])
```

```python
model.compile(
  optimizer='adam',
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])
```

```python
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=3
)
```

```python
history.history
```

```python
print (model.evaluate(train_ds, batch_size = 128))
```

```python
print (model.evaluate(val_ds, batch_size = 128))
```
